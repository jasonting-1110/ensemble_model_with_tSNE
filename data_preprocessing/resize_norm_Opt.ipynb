```python
import os
import numpy as np
import matplotlib.pyplot as plt
import copy
import time
from pathlib import Path
from scipy.ndimage.interpolation import zoom




#讀取dat轉換為npy的data，經過resize後轉npy png，再存到該目錄
# 獲取目錄中的子資料夾列表
path_file = r'E:\OCT\dental OCT\bare tooth\ensemble_model_aug\code\2024_8_13\nor_raw'
subfolders = sorted( [f for f in os.listdir(path_file) if os.path.isdir(os.path.join(path_file, f))] )
# 獲取目錄中的子資料夾列表
subfolders = sorted( [f for f in os.listdir(path_file) if os.path.isdir(os.path.join(path_file, f))] )

# 按子資料夾名稱中的數字部分進行排序: 因為python的listdir排序不完全依照數字大小排 ex:14 15 2 25 3 37...
subfolders_sorted = sorted(subfolders, key=lambda x: int(x[1:]))

# 选择指定子文件夹 : 改index決定要讀取哪些subfolder
selected_subfolders = subfolders_sorted[0:1] #必須用slice指定要處理的folders

for subfolder in selected_subfolders: 

    subfolder_path = os.path.join(path_file, subfolder)
    subfolder_npy = os.path.join(subfolder_path, 'npy')

    subfolder_resize = os.path.join(subfolder_path, 'npy_resize') 
    npy_save = os.path.join(subfolder_resize, 'npy')
    png_save = os.path.join(subfolder_resize, 'png')

    # print(f"Processing folder: {subfolder_path}") #n1 n2 ...n24
    # print(f"Processing folder: {subfolder_npy}")
    # print(f"Processing folder: {subfolder_resize}")
    # print(f"Processing folder: {npy_save}")
    # print(f"Processing folder: {png_save}")

    if Path(subfolder_resize).exists() is False:
            os.makedirs(subfolder_resize)        
    if Path(npy_save).exists() is False:
            os.makedirs(npy_save) 
    if Path(png_save).exists() is False:
            os.makedirs(png_save)       
    
    for file_num, file_name in enumerate(os.listdir(subfolder_npy)):
        if 'npy' in file_name:
            if file_num%50 == 0:
                print('Folder {} , Have processed: {}, total: {}.'.format(subfolder_npy,file_num, len(os.listdir(subfolder_npy))))
            # filename = os.path.join(root, f)          
            Bscan=np.load(subfolder_npy + '/'+ file_name)
            image_resize = zoom(Bscan,[0.407,0.112], prefilter=False)   # 原圖大小(550,2000) => (224,224)  根據比例調整[]中數字     
            
            
            # normalize image from 0 to 1
            #image_resize = (image_resize - np.min(image_resize))/(np.max(image_resize) - np.min(image_resize))
            plt.imsave(png_save + '/frame%d.png' %(file_num+1), image_resize, cmap='gray')
            np.save(npy_save + '/frame%d.npy' %(file_num+1), image_resize, allow_pickle=False)


#許多圖像大小不一，造成error


#d槽爆掉!!!
```
